```python
def ConstBoard(board):                                        #it prints the current state of the board(snapshot)
  print("Current State of the Board: \n\n");
  for i in range(0, 9):
    if((i>0) and (i%3==0)):
      print("\n");
    if(board[i]==0):
        print("_", end=" ");
    if(board[i]==-1):
        print("X", end=" ");
    if(board[i]==1):
        print("O", end=" ");
  print("\n\n");

def User1Turn(board):
  pos = input("Enter X's position from [1,2,3......,9]");   # X position
  pos = int(pos);
  if(board[pos-1]!=0):
    print("Wrong Move");
    exit(0);                                                #stops the game/step
  board[pos-1]=-1;                                          #-1 stands for X

def User2Turn(board):
  pos = input("Enter O's position from [1,2,3......,9]");   # O position
  pos = int(pos);
  if(board[pos-1]!=0):
    print("Wrong Move");
    exit(0);                                                #stops the game/step
  board[pos-1]=1;                                           #1 stands for O

def minmax(board, player):
  x = analyzeboard(board);
  if(x!=0):
    return (x*player);                                      #it means the previous player has won
  pos = -1;                                                 #we have three possibilities as 0(draw),1(O),-1(X)
  value = -2;
  for i in range(0, 9):
    if (board[i]==0):
      board[i]=player;
      score=-minmax(board, player*-1);                     #it reverses the step of minmax algo     #player*-1 gives the next player the chance to play
      board[i] = 0;
      if(score>value):
        value = score;
        pos=i;
  if(pos==-1):                                            #if there is no place to put the number/play the chance
    return 0;                                             #return Draw
  return value;                                           #or else return the value which is greatest



def CompTurn(board):                                        #we need 1 for AI to win
  pos = -1;                                                 #we have three possibilities as 0(draw),1(O),-1(X)
  value = -2;                                               #we have taken a lowest number to compare with the currentscore
  for i in range(0, 9):
    if (board[i]==0):
      board[i]=1;
      score=-minmax(board, -1);
      board[i] = 0;
      if(score>value):                                    #compare the currentscore with the lowest score -2 and the greater value will be the currentscore
        value = score;
        pos=i;
  board[pos]=1






def analyzeboard(board):
  cb = [[0, 1, 2], [3, 4, 5], [6, 7, 8], [0, 3, 6], [1, 4, 7], [2, 5, 8], [0, 4, 8], [2, 4, 6]]

  for i in range(0, 8):
    if(board[cb[i][0]]!=0 and                            #2 dimensional list
       board[cb[i][0]]==board[cb[i][1]] and              #place0 = place1 & place0 = place2
       board[cb[i][0]]==board[cb[i][2]]):
       return board[cb[i][0]];                           #return the list which is same

  return 0;                                              #or else return Drawt






def main():
  choice = input("Enter 1 for Single Player, 2 for Multiplayer:");
  choice = int(choice);
  board = [0, 0, 0, 0, 0, 0, 0, 0, 0]             #numerically represented my game board empty
  if(choice==1):
    print("Computer: 0 Vs. You: X");
    player = input("Enter to play 1(st) or 2(nd): ");
    player = int(player);
    for i in range(0, 9):                                              #for loop for Single Player
      if(analyzeboard(board)!=0):
        break;
      if((i+player)%2==0):
        CompTurn(board);
      else:
        ConstBoard(board);
        User1Turn(board);
  else:                                                                 #for loop for Multiplayer
    for i in range(0, 9):
      if(analyzeboard(board)!=0):
        break;
      if(i%2==0):
        ConstBoard(board);
        User1Turn(board);
      else:
        ConstBoard(board);
        User2Turn(board);

  x = analyzeboard(board);                                             #it analyzes the gameboard either somebody has won or is it another player chance
  if(x==0):
    ConstBoard(board);
    print("Draw!");
  if(x==-1):
    ConstBoard(board);
    print("Player X Wins!!!! O Looses");
  if(x==1):
    ConstBoard(board);
    print("Player O Wins!!!! X Looses");

```
